```python
from dotenv import load_dotenv
import os
from anthropic import Anthropic
```

```python
## load the environment variables from the .env file

load_dotenv()
```

```python
## create an instance of the Anthropic client
client = Anthropic()

model = "claude-sonnet-4-0"  # or "claude-instant-100k"
```

```python
def add_user_message(messages, content):
    messages.append({"role": "user", "content": content})


def add_assistant_message(messages, content):
    messages.append({"role": "assistant", "content": content})
    

## make a request to the API

def chat_with_claude(messages): 
    response = client.messages.create(
        model=model,
        max_tokens=1000,
        messages=messages
    )
    return response.content[0].text
```

```python
## make a list of messages 
messages = []

## add innitial user message
add_user_message(messages, "What is the meaning of life?")

messages

## pass list of messages to the chat function and print the response
response = chat_with_claude(messages)

## take message and add it as an assistant message to the list of messages
add_assistant_message(messages, response)

## add to user follow up question
add_user_message(messages, "What is the airspeed of a laden swallow?")

## call the chat function again with the updated list of messages and print the response
response = chat_with_claude(messages)
```

```python

def main():
    ## Make initial list of all messages 
    messages = []

    while True: 
        # get user input
        user_input = input("User: ")
        print('>', user_input)
        # add user message to list of messages
        add_user_message(messages, user_input)
        # get response from claude
        response = chat_with_claude(messages)
        # add response to list of messages
        add_assistant_message(messages, response)
        # print response
        print('-------------------------------------------------')
        print("Claude:", response) 
        print('-------------------------------------------------')
```

```python
main()
```
